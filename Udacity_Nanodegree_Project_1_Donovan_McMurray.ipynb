```python
def linear_regression(features, values):
    """
    Perform linear regression given a data set with an arbitrary number of features.
    
    This can be the same code as in the lesson #3 exercise.
    """
    
    features = sm.add_constant(features)
    model = sm.OLS(values, features)
    results = model.fit()
    intercept = results.params[0]
    params = results.params[1:]
    
    return intercept, params

def predictions(dataframe, feature_list, test_data=None):
    
    if test_data is None:
        test_data = dataframe
        
    add_dummy_units, add_dummy_conds = False, False
    if 'UNIT' in feature_list:
        feature_list = filter(lambda x: x != 'UNIT', feature_list)
        add_dummy_units = True
    if 'conds' in feature_list:
        feature_list = filter(lambda x: x != 'conds', feature_list)
        add_dummy_conds = True
        
    features = dataframe[feature_list]
    test_data_features = test_data[feature_list]
    
    if add_dummy_units:
        dummy_units = pandas.get_dummies(dataframe['UNIT'], prefix='unit')
        features = features.join(dummy_units)
        test_dummy_units = pandas.get_dummies(test_data['UNIT'], prefix='unit')
        test_data_features = test_data_features.join(test_dummy_units)

    if add_dummy_conds:
        dummy_conds = pandas.get_dummies(dataframe['conds'], prefix='conds')
        features = features.join(dummy_conds)
        test_dummy_conds = pandas.get_dummies(test_data['conds'], prefix='conds')
        test_data_features = test_data_features.join(test_dummy_conds)
    
    # Values
    values = dataframe['ENTRIESn_hourly']

    # Perform linear regression
    intercept, params = linear_regression(features, values)
    
    predictions = intercept + np.dot(test_data_features, params)
    return predictions

def calculate_r_squared(data, predictions):
    ss_tot = sum((data - np.mean(data)) ** 2)
    ss_res = sum((data - predictions) ** 2)
    r_squared = 1 - ss_res / ss_tot
    return r_squared

def feature_importances(input_data, outcomes):
    forest = ExtraTreesClassifier(n_estimators=30, random_state=0)
    forest.fit(input_data, outcomes)
    return forest.feature_importances_
```

```python
p = predictions(data, ['rain', 'precipi', 'UNIT'])
```

```python
calculate_r_squared(data['ENTRIESn_hourly'], p)
```

```python
blacklist_features = ['DATEn', 'UNIT', 'conds', 'TIMEn', 'ENTRIESn', 'EXITSn', 'ENTRIESn_hourly', 'EXITSn_hourly', 'datetime', 'station']
all_features = filter(lambda x: x not in blacklist_features, data.columns.values)
all_features
```

```python
sample_percentage = 0.2
sample_count = int(sample_percentage * len(data))
importances = feature_importances(data[all_features].sample(sample_count), data['ENTRIESn_hourly'].sample(sample_count))
ranked_features = list(reversed(sorted(zip(all_features, importances), key=lambda x: x[1])))
ranked_features
```

```python
important_features = [x[0] for x in ranked_features if x[1] > 0.04]
important_features
```

```python
p = predictions(data, important_features + ['UNIT', 'conds'])
calculate_r_squared(data['ENTRIESn_hourly'], p)
```

```python
p = predictions(data, all_features + ['UNIT', 'conds'])
calculate_r_squared(data['ENTRIESn_hourly'], p)
```

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(data, data['ENTRIESn_hourly'], test_size=0.4, random_state=0)
```

```python
p = predictions(X_train, important_features + ['UNIT', 'conds'], X_test)
calculate_r_squared(y_test, p)
```

```python
p = predictions(X_train, all_features + ['UNIT', 'conds'], X_test)
calculate_r_squared(y_test, p)
```
